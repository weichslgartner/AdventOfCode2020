```python
def to_list(build_output : str) -> List[str]:
    perf = []
    for line in build_output.split('\n'):
        if line.startswith('real'):
            perf.append( float(line.split()[1][2:-1].replace(',','.')))
    return perf
```

```python
clang = to_list(perf_clang)
```

```python
gcc = to_list(perf_gcc)
```

```python
days = np.arange(1,len(gcc)+1)
```

```python
width = 0.4
fig, ax = plt.subplots()
fig.set_size_inches(17, 9)
ax.set_title("Execution Time of Advent of Code 2020 C++ Solutions (Both Parts)")
rects1 = ax.bar(days - width/2, clang, width, label='clang 11')
rects2 = ax.bar(days + width/2, gcc,width, label='gcc 10.1')
ax.set_xticks(days)
ax.set_ylabel('Execution Time [s]')
ax.set_xlabel('Day')
ax.legend()
```
